```python
import json
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
def create_df_from_json_file(filename: str) -> pd.core.frame.DataFrame:
    with open(filename) as f:
        lines = f.read().split("\n")
    
    entities = []
    for line in lines:
        try:
            entities.append(json.loads(line))
        except Exception as error:
            print(error)
            continue
    df = pd.DataFrame(entities)
    df = df.set_index("id")
    return df

def create_df_for_label_file(filename: str, label_colname: str) -> pd.core.frame.DataFrame:
    with open(filename) as f:
        json_object = json.load(f)
    df = pd.DataFrame({"id": list(json_object.keys()), label_colname: list(json_object.values())})
    df["id"] = df["id"].astype(int)
    df = df.set_index("id")
    return df
```

```python
df_training_set = create_df_from_json_file("2_training_set.json")
df_training_anno_top = create_df_for_label_file("2_training_anno_top.json", "label_top")
df_training_anno_mid = create_df_for_label_file("2_training_anno_mid.json", "label_mid")
df_training_anno_fine = create_df_for_label_file("2_training_anno_fine.json", "label_fine")
```
